<h2 style="color:#2e86c1; background-color:#f0f0f0; padding:10px; border-radius:10px; text-align:center;">Objetivo del Proyecto</h2>

<p style="color:#2e86c1;">
  El objetivo del proyecto es desarrollar un entorno de simulación sofisticado basado en aprendizaje por refuerzo, diseñado específicamente para entrenar agentes capaces de interactuar y adaptarse a los mercados financieros. Este entorno tiene como propósito no solo modelar escenarios de mercado históricos y actuales, sino también generar condiciones desafiantes y volátiles que pongan a prueba la capacidad de los agentes para gestionar riesgos y tomar decisiones óptimas bajo incertidumbre.
</p>

<p style="color:#2e86c1;">
  Los agentes entrenados en este entorno podrán tomar decisiones secuenciales, como comprar, vender o mantener activos, considerando factores como los cambios en los precios, las tendencias del mercado y la volatilidad esperada. El enfoque no se limita a maximizar las ganancias; también busca promover la estabilidad y la resiliencia a través de un sistema de recompensas que fomenta la preservación del capital y la gestión eficiente del riesgo.
</p>

<p style="color:#2e86c1;">
  Además, este proyecto prioriza la implementación de herramientas analíticas avanzadas para evaluar el rendimiento de las estrategias de trading. Estas herramientas proporcionarán métricas clave que permitirán identificar fortalezas, áreas de mejora y las condiciones en las que cada estrategia sobresale. En última instancia, el proyecto busca establecer una base sólida para la integración de algoritmos de trading en escenarios reales, optimizando la capacidad de adaptación y la rentabilidad en mercados financieros dinámicos y complejos.
</p>

<h2 style="color:#2e86c1; background-color:#f0f0f0; padding:10px; border-radius:10px; text-align:center;">Documentación de los Indicadores Técnicos</h2>

<p style="color:#2e86c1;">
  Los indicadores técnicos son herramientas matemáticas utilizadas en el análisis financiero para interpretar datos históricos de precios y volúmenes. Estos indicadores ayudan a identificar patrones, tendencias y posibles puntos de entrada o salida en los mercados financieros. Su relevancia radica en su capacidad para simplificar y cuantificar comportamientos complejos del mercado, proporcionando a los traders y algoritmos información clave para tomar decisiones informadas. En el contexto del aprendizaje por refuerzo, los indicadores técnicos permiten a los agentes evaluar las condiciones del mercado de manera dinámica, integrando señales como impulso, volatilidad y niveles de sobrecompra o sobreventa para optimizar sus estrategias de trading. Esto los convierte en una pieza fundamental para modelar y predecir el comportamiento de los activos financieros.
</p>

<h2 style="color:#2e86c1; background-color:#f0f0f0; padding:10px; border-radius:10px; text-align:center;">Descripción de cada indicador</h2>

<h3 style="color:#2e86c1;">1. Relative Strength Index (RSI)</h3>
<p style="color:#2e86c1;">
  El <strong>RSI</strong> (Índice de Fuerza Relativa) es un oscilador técnico que evalúa la fuerza de los movimientos de precios recientes para determinar si un activo está sobrecomprado o sobrevendido. Fue desarrollado por J. Welles Wilder en 1978 y es ampliamente utilizado en el análisis técnico.
</p>
<p style="color:#2e86c1;">
  El RSI oscila entre 0 y 100 y se calcula utilizando la media de las ganancias y pérdidas durante un período definido, típicamente 14 días. Los valores altos (por encima de 70) indican sobrecompra, mientras que los valores bajos (por debajo de 30) sugieren sobreventa.
</p>
<ul style="color:#2e86c1;">
  <li><strong>Cálculo:</strong> </li>
    <ul>
    <li>Compra: RSI = 100 - (100 / (1 + RS)), donde RS es la relación entre la ganancia promedio y la pérdida promedio.</li>
  </ul>
  <li><strong>Señales:</strong></li>
  <ul>
    <li>Compra: Cuando el RSI cae por debajo de un umbral, en este proyecto 94.5 (señal de sobreventa).</li>
    <li>Venta: Cuando el RSI supera un umbral, aquí 98.5 (señal de sobrecompra).</li>
  </ul>
</ul>


<h3 style="color:#2e86c1;">2. Bollinger Bands</h3>
<p style="color:#2e86c1;">
  Las <strong>Bollinger Bands</strong> son un indicador de volatilidad desarrollado por John Bollinger en los años 80. Consisten en tres líneas: una media móvil simple (SMA) y dos bandas que representan una desviación estándar por encima y por debajo de la SMA. Estas bandas se expanden o contraen en función de la volatilidad del mercado.
</p>
<ul style="color:#2e86c1;">
  <li><strong>Cálculo:</strong></li>
  <ul>
    <li>Banda Superior: SMA + (k * desviación estándar)</li>
    <li>Banda Inferior: SMA - (k * desviación estándar)</li>
  </ul>
  <li><strong>Señales:</strong></li>
  <ul>
    <li>Compra: Cuando el precio cae por debajo de la banda inferior, indicando posible reversión al alza.</li>
    <li>Venta: Cuando el precio supera la banda superior, señal de posible reversión a la baja.</li>
  </ul>
</ul>



<h3 style="color:#2e86c1;">3. Moving Average Convergence Divergence (MACD)</h3>
<p style="color:#2e86c1;">
  El <strong>MACD</strong> es un indicador de impulso que mide la relación entre dos medias móviles exponenciales (EMA): una rápida (de menor período) y una lenta (de mayor período). Fue desarrollado por Gerald Appel en la década de 1970.
</p>
<ul style="color:#2e86c1;">
  <li><strong>Cálculo:</strong></li>
  <ul>
    <li>Línea MACD: EMA rápida - EMA lenta.</li>
    <li>Línea de Señal: Media móvil exponencial de la línea MACD.</li>
  </ul>
  <li><strong>Señales:</strong></li>
  <ul>
    <li>Compra: Cuando la línea MACD cruza hacia arriba la línea de señal.</li>
    <li>Venta: Cuando la línea MACD cruza hacia abajo la línea de señal.</li>
  </ul>
</ul>

<h2 style="color:#2e86c1; background-color:#f0f0f0; padding:10px; border-radius:10px; text-align:center;">Combinación de Indicadores</h2>


<p style="color:#2e86c1; margin-bottom: 20px;">
  La combinación de <strong>RSI</strong>, <strong>Bollinger Bands</strong>, y <strong>MACD</strong> fue determinada tras un análisis riguroso basado en pruebas simuladas con datos históricos y escenarios generados. Los parámetros seleccionados para cada indicador fueron optimizados para maximizar la sensibilidad a los cambios de mercado y la generación de señales útiles, minimizando al mismo tiempo las señales falsas. Este enfoque buscó equilibrar precisión, adaptabilidad y robustez en distintos contextos de mercado.
</p>

<ul style="color:#2e86c1; margin-bottom: 20px;">
  <li><strong>RSI:</strong> Este indicador se configuró con ventanas ajustadas a valores de 94.5 y 98.5. Estas configuraciones, lejos de los valores tradicionales (70 y 30), fueron seleccionadas debido a que las pruebas mostraron una mayor eficacia al detectar fluctuaciones menores en escenarios de alta actividad. La alta sensibilidad del RSI permitió identificar condiciones de sobrecompra y sobreventa con más frecuencia, lo que resultó en un mayor número de señales operables para el agente, ajustándose así al objetivo de maximizar las oportunidades de trading en el horizonte temporal estudiado.</li>
  
  <li><strong>Bollinger Bands:</strong> Se emplearon ventanas de 2 días y una desviación estándar de 0.5 para ajustar las bandas a movimientos rápidos en el mercado. Las pruebas indicaron que esta configuración optimizada permitía capturar mejor los cambios de volatilidad a corto plazo, haciendo de este indicador una herramienta crítica en mercados altamente dinámicos. Estas bandas ajustadas ofrecieron un marco claro para identificar puntos de reversión temprana y aprovecharlos de manera efectiva.</li>
  
  <li><strong>MACD:</strong> La selección de ventanas de 3 y 6 días para las medias móviles exponenciales rápida y lenta respectivamente se basó en la necesidad de captar señales de impulso en períodos cortos, sin perder la capacidad de confirmar tendencias emergentes. Esta configuración redujo el tiempo de reacción del indicador, mejorando su capacidad para sincronizarse con las señales del RSI y las Bollinger Bands. Además, se maximizó la utilidad del MACD como un filtro para validar señales de compra o venta generadas por los otros indicadores.</li>
</ul>

<p style="color:#2e86c1; margin-bottom: 20px;">
  La elección de estos parámetros respondió a un análisis de sensibilidad que evaluó múltiples configuraciones en distintos escenarios de mercado. Se observó que esta combinación ofrecía un equilibrio óptimo entre la generación de señales frecuentes y la reducción de ruido en las mismas. Además, permitió que el agente de trading operara de manera más efectiva en contextos de alta y baja volatilidad, adaptándose dinámicamente a las condiciones cambiantes del mercado.
</p>

<p style="color:#2e86c1;">
  En conclusión, la integración de estos indicadores y la calibración de sus parámetros específicos resultaron en un sistema que optimizó la detección de oportunidades, mejoró la precisión en la toma de decisiones y aseguró la robustez del modelo frente a diferentes escenarios de mercado. Este enfoque demuestra cómo el diseño meticuloso de parámetros puede incrementar significativamente la efectividad de las estrategias basadas en aprendizaje por refuerzo.
</p>

<h2 style="color:#2e86c1; background-color:#f0f0f0; padding:10px; border-radius:10px; text-align:center;">Combinación de Indicadores</h2>


<p style="color:#2e86c1; margin-bottom: 20px;">
La clase step es una función clave que gestiona cómo evoluciona el entorno de trading en respuesta a las decisiones del agente. Este método simula las operaciones de compra, venta y mantenimiento de posiciones en un entorno financiero basado en precios históricos y señales de indicadores técnicos.

<p style="color:#2e86c1; margin-bottom: 20px;">
Propósito:
El método actualiza el estado del entorno, calcula la recompensa en función de la acción tomada, y determina si la simulación ha llegado a su fin. También evalúa el impacto de la acción del agente en su patrimonio neto (net_worth) y en las señales técnicas disponibles.

<p style="color:#2e86c1; margin-bottom: 20px;">
Funcionamiento General
Inicialización de Variables Se identifica el precio actual del activo en el paso temporal correspondiente y se evalúan las señales de los indicadores técnicos (MACD, RSI y Bandas de Bollinger). Estas señales determinan si el mercado sugiere comprar, vender o mantener la posición.

<p style="color:#2e86c1; margin-bottom: 20px;">
Acción del Agente
        
<p style="color:#2e86c1; margin-bottom: 20px;">
Comprar: Si el agente decide comprar, se verifica que tenga suficiente balance disponible y que no exceda el límite máximo de posiciones. Al comprar, se deduce el costo de la transacción (incluyendo comisiones) de su balance. Además, el agente recibe una recompensa adicional si el precio sube en el siguiente paso temporal.
Vender: Si el agente opta por vender, se verifica que tenga suficientes unidades disponibles para vender. Al vender, el balance aumenta proporcionalmente, descontando las comisiones. Si el precio baja en el siguiente paso, el agente es recompensado.
Mantener: Si el agente decide mantener su posición, no realiza ninguna operación. Sin embargo, si hay señales técnicas claras que sugieren actuar y el agente decide mantenerse inactivo, recibe una penalización leve.
Bonificaciones por Indicadores Técnicos Se otorgan recompensas adicionales si las acciones del agente están alineadas con las señales de los indicadores técnicos:

<p style="color:#2e86c1; margin-bottom: 20px;">    
MACD: Bonificación por seguir señales de compra o venta.
RSI: Bonificación por actuar en situaciones de sobrecompra o sobreventa.
Bandas de Bollinger: Bonificación por operar en niveles de soporte o resistencia.
Cálculo del Patrimonio Neto El patrimonio neto del agente se actualiza sumando su balance disponible y el valor de sus posiciones actuales. Si el patrimonio neto disminuye como resultado de la acción, el agente recibe una penalización proporcional a la pérdida.

<p style="color:#2e86c1; margin-bottom: 20px;">    
Condición de Finalización La simulación concluye si el agente alcanza el último paso temporal de los datos de precios. En ese caso, el entorno marca la simulación como completada (done = True).

<p style="color:#2e86c1; margin-bottom: 20px;">
Actualización del Estado El método genera el siguiente estado observable para el agente, el cual incluye información relevante como los precios, indicadores técnicos y otras métricas necesarias para tomar decisiones en el siguiente paso.

<p style="color:#2e86c1; margin-bottom: 20px;">    
Recompensas y Penalizaciones
El sistema de recompensas está diseñado para incentivar acciones que maximicen el patrimonio neto del agente y alinearse con señales técnicas del mercado. Las penalizaciones buscan desalentar decisiones que resulten en pérdidas o inactividad frente a señales claras.

<p style="color:#2e86c1; margin-bottom: 20px;">    
Resultados
Al final del método, se devuelve:

<p style="color:#2e86c1; margin-bottom: 20px;">    
El nuevo estado observable del entorno.
La recompensa obtenida en el paso actual.
Si la simulación ha concluido.
Información adicional que puede ser útil para el agente o el análisis posterior.
Este enfoque permite que el agente aprenda estrategias de trading óptimas mediante la interacción con el entorno y la retroalimentación obtenida a través de las recompensas y penalizaciones.
</p>
